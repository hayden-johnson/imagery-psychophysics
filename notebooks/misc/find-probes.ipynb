```python
path_to_curves = '../../results/curves/'

curves = []
for i in range(12):
    # load in curves (saved as txt files)
    curve = np.loadtxt(path_to_curves + 'curve_' + str(i) + '.txt')
    curves.append(curve)
```

```python
from curves.probes import generate_custom_probe_set_ext
```

```python
probes = generate_custom_probe_set_ext(2, ext=.2)
```

```python
!mkdir ./probes
```

```python
probes_path = './probes/'
# save probes to txt file
for i, probe in enumerate(probes):
    np.savetxt(probes_path + 'probe_' + str(i) + '.txt', probe)

# load in from txt files
probes_loaded = []
for i in range(len(probes)):
    probe = np.loadtxt(probes_path + 'probe_' + str(i) + '.txt')
    probes_loaded.append(probe)
```

```python
probes = probes_loaded
```

```python
for tid in range(len(curves)):
    for pid in range(len(probes)):
        if tid != pid:
            image_path = f'../../results/resources/isecs_{tid}_{pid}.png'
            fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(10, 4))
            img = plt.imread(image_path)
            ax[0].imshow(img)
            pretty_picture_of_curve_ax(ax[1], [curves[tid], probes[pid]], color=['black', 'deeppink'])
            plt.show()
```
